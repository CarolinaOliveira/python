```python
#Webdriver é um livro da biblioteca do selenium por isso uso o from 
#Selenium roda em tread, o pyautogui não.



from selenium import webdriver #controlar o navegador
from selenium.webdriver.common.keys import Keys #controlar as teclas do teclado
from selenium.webdriver.common.by import By #localizar itens do navegador

#crira navegador
navegador = webdriver.Chrome()

# entrar no google pesquisar cotação do dolar
navegador.get("https://www.google.com.")

#pegar cotação dolar #para utilização do XPATH utilize aspas simples 

navegador.find_element(By.XPATH, '/html/body/div[1]/div[3]/form/div[1]/div[1]/div[1]/div/div[2]/input').send_keys("cotação dólar",Keys.ENTER) #qual o parametro vc quer usar, qual o código do parametro)

#navegador.find_element(By.XPATH, '/html/body/div[1]/div[3]/form/div[1]/div[1]/div[1]/div/div[2]/input').send_keys(Keys.ENTER)

cotacao_dolar = navegador.find_element(By.XPATH,'//*[@id="knowledge-currency__updatable-data-column"]/div[1]/div[2]/span[1]').get_attribute('data-value')
print(cotacao_dolar)


# entrar no google pesquisar cotação do euro

navegador.get("https://www.google.com.")
navegador.find_element(By.XPATH, '/html/body/div[1]/div[3]/form/div[1]/div[1]/div[1]/div/div[2]/input').send_keys("cotação Euro",Keys.ENTER) #qual o parametro vc quer usar, qual o código do parametro)

#pegar cotação Euro
cotacao_euro = navegador.find_element(By.XPATH,'//*[@id="knowledge-currency__updatable-data-column"]/div[1]/div[2]/span[1]').get_attribute('data-value')
print(cotacao_euro)


# entrar no google pesquisar cotação do ouro
navegador.get('https://www.melhorcambio.com/ouro-hoje')

#pegar cotação do ouro 
cotacao_ouro = navegador.find_element(By.XPATH,'//*[@id="comercial"]').get_attribute('value')

#substituir vírgula por ponto
cotacao_ouro = cotacao_ouro.replace(",", ".")

print(cotacao_ouro)

#sair do navegador
navegador.quit()


```

```python
#importar base de dados
import pandas as pd

tabela = pd.read_excel("Produtos.xlsx") #esse arquivo precisa estar na mesma pasta desse arquivo
display(tabela)
```

```python
#atualizar cotações na base de dados 
#tabela["Moeda"]=="Dólar" => Traga os valores das linhas com nome dolar na coluna moeda


tabela.loc[tabela["Moeda"]=="Dólar", "Cotação"] = float(cotacao_dolar) # loc = localizar [linhas, colunas]

tabela.loc[tabela["Moeda"]=="Euro", "Cotação"] = float(cotacao_euro)

tabela.loc[tabela["Moeda"]=="Ouro", "Cotação"] = float(cotacao_ouro)

#atualizar o preço de compra na base de dados
tabela["Preço de Compra"] = tabela["Preço Original"] * tabela ["Cotação"]

#atualizar o preço de venda na base de dados
tabela["Preço de Venda"] = tabela["Preço de Compra"] * tabela ["Margem"]

display(tabela)
```

```python
tabela.to_excel("ProdutosNovo.xlsx", index=False)
```
